```python
from sklearn.datasets import make_classification
```

```python
X, y = make_classification(n_samples=1000, n_features=10, n_classes=3, random_state=15,n_informative=3)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
logistic = LogisticRegression()
logistic.fit(X_train, y_train)
```

```python
y_pred = logistic.predict(X_test)
```

```python
y_pred
```

```python
## Metrics

from sklearn.metrics import classification_report, accuracy_score, confusion_matrix

print("accuracy score")
print(accuracy_score(y_test, y_pred))
print("Classification report")
print(classification_report(y_test, y_pred))
print("Confusion matrix")
print(confusion_matrix(y_test, y_pred))
```

```python
## Hyper-parameter turning

from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.linear_model import LogisticRegression

stratified = StratifiedKFold(n_splits=15)
penalty = ['l1', 'l2', 'elasticnet']
c = [100, 10.0, 0.01, 0.001]
random_state = 42
solver = ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga']
multi_class = ['ovr']
n_jobs = -1

logs = LogisticRegression()
```

```python
params = dict(
    penalty = penalty,
    C = c,
    solver = solver,
    multi_class = multi_class
)
```

```python
grid = GridSearchCV(estimator = logs, param_grid = params, n_jobs=-1, cv=stratified, scoring='accuracy')
```

```python
grid.fit(X_train, y_train)
```

```python
y_pred = grid.predict(X_test)
```

```python
grid.best_score_
```

```python
grid.best_estimator_
```

```python
from sklearn.model_selection import RandomizedSearchCV
random = RandomizedSearchCV(estimator=logs, param_distributions=params, scoring='accuracy', n_jobs=-1, cv=stratified, random_state=42)
```

```python
random.fit(X_train, y_train)
```

```python
y_pred = random.predict(X_test)
```
