```python
## Remove unnecessary columns
dataset.drop(columns=['brand', 'car_name'], inplace=True)
```

```python
## Indepedent and dependent features
X = dataset.drop(columns=['selling_price'], axis=1)
y = dataset['selling_price']
```

```python
X.head()
```

```python
y.head()
```

```python
len(X['model'].unique())
```

```python
## Use label encode as it has unique of 120 values so it does not fit with one hot encoder
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
X['model'] = label.fit_transform(X['model'])
```

```python
X['model']
```

```python
## Numerical and categorical features
numerical_features = X.select_dtypes(exclude='object').columns
categorical_features = X.select_dtypes(include='object').columns
```

```python
## Column transformation (remainder ='passthrough' meaning it should 
# transform mention columns but rest be should be intact)

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder

onehot = OneHotEncoder(drop='first')
standard = StandardScaler()

preprocessor = ColumnTransformer(
    [
        ('OneHotEncoder', onehot, ['transmission_type', 'fuel_type', 'seller_type']),
        ('StandardScaler', standard, numerical_features)
    ], remainder='passthrough'
)
```

```python
preprocessor
```

```python
X = preprocessor.fit_transform(X)
```

```python
pd.DataFrame(X)
```

```python
## Train and test split

from sklearn.model_selection import train_test_split
X_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=32)
```

```python
## Training a model

from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
regressor = RandomForestRegressor()
regressor.fit(X_train, y_train)
```

```python
y_pred =regressor.predict(x_test)
```

```python
## Metrics
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
print(f"Mean squared error {mean_squared_error(y_test, y_pred)}")
print(f"R2 score {r2_score(y_test, y_pred)}")
print(f"Mean absolute error {mean_absolute_error(y_test, y_pred)}")
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
models = {
    "randomforest": RandomForestRegressor(),
    "linear": LinearRegression(),
    "KNeighborsRegressor": KNeighborsRegressor(),
    "DecisionTreeRegressor": DecisionTreeRegressor(),
    "Lasso": Lasso(),
    "Ridge": Ridge(),
    "Gradient": GradientBoostingRegressor()
}
```

```python
import numpy as np
for i in range(len(list(models))):
    
    model_name = list(models.values())[i]
    print(f"Model name - {list(models.keys())[i]}")
    print(f"*******************************")
    model_name.fit(X_train, y_train)
    y_train = model_name.predict(X_train)
    y_pred = model_name.predict(x_test)
    print(f"r2_score {r2_score(y_train, y_train)}")
    print(f"Mean squared error {mean_squared_error(y_train, y_train)}")
    print(f"Mean Absolute error {mean_absolute_error(y_train, y_train)}")
    print(f"Root mean squared error {np.sqrt(mean_squared_error(y_train, y_train))}")
    
    print(f"r2_score {r2_score(y_test, y_pred)}")
    print(f"Mean squared error {mean_squared_error(y_test, y_pred)}")
    print(f"Mean Absolute error {mean_absolute_error(y_test, y_pred)}")
    print(f"Root mean squared error {np.sqrt(mean_squared_error(y_test, y_pred))}")
    print(f"*******************************")
    
```

```python
## Hyperparameter turning
rfr = {
    "max_depth": [5, 8, 15, None, 10],
    "max_features": [5, 7, "auto", 8],
    "min_samples_split": [2, 8, 15, 20],
    "n_estimators": [100,200,500,1000],
}

gradient = {
   "loss" : ['squared_error', 'absolute_error', 'huber','quantile'],
   "learning_rate": [0.1, 0.01, 0.02, 0.03],
   "n_estimators": [100,200,300,500],
   "criterion": ['friedman_mse', 'squared_error'],
   "min_samples_split": [2, 8, 15, 20],
   "max_depth": [5, 8, 15, None, 10],
}

models = [
    ('RandomForest', RandomForestRegressor(), rfr),
    ('Gradient', GradientBoostingRegressor(), gradient)
]
```

```python
from sklearn.model_selection import RandomizedSearchCV

for model_name, model, param in models:
    
    print(f"Model Name {model_name}")
    random = RandomizedSearchCV(model, param_distributions=param, n_iter=100, cv=3, verbose=2, n_jobs=1)
    random.fit(X_train, y_train)
    print(random.best_estimator_)
    print(random.best_score_)
```
