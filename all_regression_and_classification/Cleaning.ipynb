```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
## Read a data from csv
data = pd.read_csv("algerian_forest_fires.csv", header=1)
data.head()
```

```python
## See the datatypes of the each features
data.info()
```

```python
## Check if there is any missing values
data.isnull().sum()
```

```python
data.isnull().any(axis=1)
```

```python
data[data.isnull().any(axis=1)]
```

```python
## Split two regions by assign 0 and 1 values to the new column called as region
data.loc[:122, "Region"] = 0
data.loc[122:, "Region"] = 1
```

```python
data.head()
```

```python
data.info()
```

```python
data = data.dropna().reset_index(drop=True)
```

```python
data[data.isnull().any(axis=1)]
```

```python
data.isnull().sum()
```

```python
### Change the datatype of region from float to int
data[['Region']] = data[['Region']].astype(int)
```

```python
data.info()
```

```python
data.iloc[[122]]
```

```python
data = data.drop(122).reset_index(drop=True)
```

```python
data.iloc[[122]]
```

```python
## Fix spaces in column names
data.columns = data.columns.str.strip()
```

```python
## Change datatype to integer
data[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']] = data[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
data.info()
```

```python
## Change the rest of the columns to integer expect the Classes column
existing_columns = [feature for feature in data.columns if data[feature].dtypes == 'O']
```

```python
existing_columns
```

```python
for i in existing_columns:
    if i != 'Classes':
        data[i] = data[i].astype(float)
```

```python
data.info()
```

```python
## Save the cleaned dataset
data.to_csv('cleaned_dataset.csv', index=False)
```
