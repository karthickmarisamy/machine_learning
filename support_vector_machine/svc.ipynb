```python
import pandas as pd
import seaborn as sns
```

```python
from sklearn.datasets import make_classification
```

```python
X, y = make_classification(n_samples = 1000, n_features = 2, n_classes = 2, n_clusters_per_class = 2, n_redundant = 0 )
```

```python
X
```

```python
y
```

```python
sns.scatterplot(data = pd.DataFrame(X), x = pd.DataFrame(X)[0], y = pd.DataFrame(X)[1], hue=y)
```

```python
## Train and test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
## Use SVM for predication with Linear
from sklearn.svm import SVC
svc =  SVC(kernel='linear')
```

```python
svc.fit(X_train, y_train)
```

```python
y_pred = svc.predict(X_test)
```

```python
## Metrics
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
## RBF Kernel

rbf = SVC(kernel = 'rbf')
rbf.fit(X_train, y_train)
```

```python
y_rbf_pred = rbf.predict(X_test)
```

```python
print(classification_report(y_test, y_rbf_pred))
print(confusion_matrix(y_test, y_rbf_pred))
```

```python
## Polynomial
poly = SVC(kernel='poly')
poly.fit(X_train, y_train)
```

```python
y_poly_pred = poly.predict(X_test)
print(classification_report(y_test, y_poly_pred))
print(confusion_matrix(y_test, y_poly_pred))
```

```python
## Sigmoid
sigmoid = SVC(kernel='sigmoid')
sigmoid.fit(X_train, y_train)
```

```python
y_sigmoid_pred = sigmoid.predict(X_test)
```

```python
print(classification_report(y_test, y_sigmoid_pred))
print(confusion_matrix(y_test, y_sigmoid_pred))
```

```python
## Hyerparameter turning

from sklearn.model_selection import GridSearchCV
params = {
    'C' : [0.1, 1, 10, 100, 1000],
    'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
    'kernel': ['rbf']
}

```

```python
grid = GridSearchCV(SVC(), param_grid=params, cv=5, verbose=3, refit=True)
```

```python
grid.fit(X_train, y_train)
```

```python
y_grid_pred = grid.predict(X_test)
```

```python
print(classification_report(y_test, y_grid_pred))
```
