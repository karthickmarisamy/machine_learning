```python
## Fill na values with median, mode

data.Age.fillna(data.Age.median(), inplace=True)
data.TypeofContact.fillna(data.TypeofContact.mode()[0], inplace=True)
data.DurationOfPitch.fillna(data.DurationOfPitch.median(), inplace=True)
data.NumberOfFollowups.fillna(data.NumberOfFollowups.mode()[0], inplace=True)
data.PreferredPropertyStar.fillna(data.PreferredPropertyStar.mode()[0], inplace=True)
data.NumberOfTrips.fillna(data.NumberOfTrips.median(), inplace=True)
data.NumberOfChildrenVisiting.fillna(data.NumberOfChildrenVisiting.mode()[0], inplace=True)
data.MonthlyIncome.fillna(data.MonthlyIncome.median(), inplace=True)
```

```python
data.drop(columns='CustomerID', axis=1, inplace=True)
```

```python
## Reduce the feature by adding NumberOfPersonVisiting and NumberOfChildrenVisiting
data['TotalVisiting'] = data['NumberOfChildrenVisiting'] + data['NumberOfPersonVisiting']
data.drop(columns=['NumberOfChildrenVisiting', 'NumberOfPersonVisiting'], inplace=True)
```

```python
data.head()
```

```python
## Get all Numerical features
num_features = [feature for feature in data.columns if data[feature].dtype != 'O']
```

```python
## Get all Numerical features
category_features = [feature for feature in data.columns if data[feature].dtype == 'O']
```

```python
## Discreate features (which is fixed size of categories like 10 or 20.. like zipcode)
discrete_features = [feature for feature in num_features if len(data[feature].unique())<=25]
```

```python
## Continoues features
continoues_features = [feature for feature in num_features if feature not in discrete_features]
```

```python
print(f"Total Numerical features {len(num_features)}")
print(f"Total Categorical features {len(category_features)}")
print(f"Total Discrete features {len(discrete_features)}")
print(f"Total Continoues features {len(continoues_features)}")
```

```python
## Train test split and model training

from sklearn.model_selection import train_test_split
X = data.drop(columns='ProdTaken', axis=1)
y = data['ProdTaken']
```

```python
X.head()
```

```python
y.value_counts()
```

```python
X_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
X_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
## Column transformation
categorical_features = X.select_dtypes(include='object').columns
numerical_features = X.select_dtypes(exclude='object').columns
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer

standard = StandardScaler()
onehot = OneHotEncoder(drop='first')
preprocess = ColumnTransformer(
    [
        ('OneHotEncoder', onehot, categorical_features),
        ('StandardScaler', standard, numerical_features)
    ]
)
```

```python
preprocess
```

```python
X_train = preprocess.fit_transform(X_train)
x_test = preprocess.transform(x_test)
```

```python
pd.DataFrame(X_train)
```

```python
## Training

from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier 
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, mean_squared_error, r2_score, roc_auc_score,recall_score,ConfusionMatrixDisplay,precision_score,f1_score, accuracy_score
```

```python
models = {
    'Decision Tree': DecisionTreeClassifier(),
    'Random forest': RandomForestClassifier(),
    'Gradient Classifier': GradientBoostingClassifier()
}

for i  in range(len(list(models))):
    model = list(models.values())[i]
    
    model.fit(X_train, y_train)
    
    ## training and test prediction
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(x_test)
    
    print(f"Model name - {list(models.keys())[i]}")
    ## Performance of training prediction
    print(f"Training Accuracy score {accuracy_score(y_train, y_train_pred)}")
    print(f"F1 score {f1_score(y_train, y_train_pred, average='weighted')}")
    print(f"Precision score {precision_score(y_train, y_train_pred)}")
    print(f"Recall score {recall_score(y_train, y_train_pred)}")
    print(f"ROC curve score {roc_auc_score(y_train, y_train_pred)}")
    
    ## Performance of test prediction
    print(f"Training Accuracy score {accuracy_score(y_test, y_test_pred)}")
    print(f"F1 score {f1_score(y_test, y_test_pred, average='weighted')}")
    print(f"Precision score {precision_score(y_test, y_test_pred)}")
    print(f"Recall score {recall_score(y_test, y_test_pred)}")
    print(f"ROC curve score {roc_auc_score(y_test, y_test_pred)}")
    print("--------------------------------------")
```

```python
### Hyperparameter turning
randomforest_params = {
    "n_estimators": [100,200,300,400],
    "criterion" : ["gini", "entropy", "log_loss"],
    "max_depth" : [5, 7, "auto",8],
    "min_samples_split": [2,8,15,20]
}

gradientboost = {
    "loss": ['log_loss', 'exponential'],
    "n_estimators" : [100,200,300,400],
    "criterion" : ['friedman_mse', 'squared_error'],
    "max_depth": [1, 5, None, 7]
}

hyper = [
    ('RF', RandomForestClassifier(), randomforest_params),
    ('GradientBoost', GradientBoostingClassifier(), gradientboost)
    
]
```

```python
from sklearn.model_selection import RandomizedSearchCV

model_param={}
for name, model, param in hyper:
    
    rscv = RandomizedSearchCV(estimator=model, param_distributions=param, cv=3, verbose=2, n_jobs=-1)

    rscv.fit(X_train, y_train)
    
    model_param[name] = rscv.best_params_
    model_param[name+'score'] = rscv.score
    
```
